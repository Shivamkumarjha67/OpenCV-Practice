```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
# By default cv2 library read image in GBR format
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Converting the image to HSV format, which is more closer to Human vision type
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
# In HLS(Hue, Saturation, Lightness) format
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
