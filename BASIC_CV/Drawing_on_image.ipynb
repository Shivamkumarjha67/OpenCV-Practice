```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
black_image = np.zeros(shape=(512, 512, 3), dtype=np.int16)
black_image.shape
```

```python
plt.imshow(black_image)
```

# DRAWING THE SHAPES ON GIVEN IMAGE

```python
# Drawing a rectangle on a already black created image.
cv2.rectangle(black_image, pt1=(350, 10), pt2=(500, 150), color=(0, 255, 0), thickness=10)
```

```python
plt.imshow(black_image)
```

```python
# Creating a another rectangle
cv2.rectangle(black_image, pt1=(300, 200), pt2=(200, 300), color=(0, 0, 255), thickness=10)
```

```python
plt.imshow(black_image)
```

```python
# Creating one more rectangle - red one
cv2.rectangle(black_image, pt1=(150, 350), pt2=(10, 500), color=(255, 0, 0), thickness=10)
```

```python
plt.imshow(black_image)
```

```python
# Drawing a circle, having some part of all rectangle
cv2.circle(black_image, center=(250, 250), radius=200, color=(100, 150, 100), thickness=8)
plt.imshow(black_image)
```

```python
# Drawing the solid circle, by manipulating the thickness
cv2.circle(black_image, center=(450, 450), radius=50, color=(100, 0, 0), thickness=-2)
plt.imshow(black_image)
```

```python
# Drawing a line as diagonal
cv2.line(black_image, pt1=(0, 0), pt2=(512, 512), color=(255, 100, 100), thickness=5)
plt.imshow(black_image)
```

# WRITING TEXT ON IMAGE

```python
font = cv2.FONT_HERSHEY_SIMPLEX    # choosing the font
cv2.putText(black_image, text='Trial', org=(200, 500), fontFace=font, fontScale=2, color=(255, 255, 255), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(black_image)
```

# DRAWING A POLYGONS

```python
black_image2 = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(black_image2)
```

```python
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
```

```python
vertices.shape
```

```python
pts.shape
```

```python
pts
```

```python
cv2.polylines(black_image2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_image2)
```
